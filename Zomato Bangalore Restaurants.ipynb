```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading CSV file

df = pd.read_csv("/Users/hemanthnarlasubramanyam/Documents/Projects/Zomato Bangalore Restaurants/zomato.csv")

df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis =1)
```

```python
df.head()
```

```python
df.info()
```

```python
#Removing the Duplicates

df.drop_duplicates(inplace = True)
```

```python
df.shape
```

```python
#cleaning the rate column

df['rate'].unique()
```

```python
#removing unwanted items in rate coulumn

def handlerate(value):
    if (value=='NEW' or value== '-'):
        return np.nan
    
    else:
        
        value = str(value).split('/')
        value = value[0]
        return float(value)
```

```python
df['rate'] = df['rate'].apply(handlerate)
```

```python
df['rate'].head()
```

```python
#filling null values in rate column with mean

df['rate'].fillna(df['rate'].mean(), inplace = True )
```

```python
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace = True)
```

```python
df.head()
```

```python
df.rename(columns = {'approx_cost(for two people)' : 'Cost2plates' , 'listed_in(type)': 'Types' ,'listed_in(city)' : 'City' } , inplace = True)

df.head()
```

```python
df['location'].unique()
```

```python
df['City'].unique()
```

```python
df = df.drop(['City'] , axis = 1)
```

```python
df['Cost2plates'].unique()
```

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',' , '')
        return float(value)
    
    else:
        return float(value)
```

```python
df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
```

```python
df['Cost2plates'].unique()
```

```python
df.head()
```

```python
#cleaning Rest Type column

rest_type = df['rest_type'].value_counts(ascending = False)

rest_type
```

```python
rest_types_lessthan1000 = rest_type[rest_type < 1000]

rest_types_lessthan1000
```

```python
#making rest type less than 1000 in frequency as "others"

def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
        
    else:
        return value
```

```python
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
```

```python
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
location = df['location'].value_counts(ascending  = False)

location_lessthan300 = location[location<300]



def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

```python
cuisines = df['cuisines'].value_counts(ascending  = False)


cuisines_lessthan100 = cuisines[cuisines<100]



def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
df.head()
```

```python
df['Types'].value_counts()
```
